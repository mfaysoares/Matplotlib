#**Tutorial Matplotlib**:



```python
import pandas as pd
import datetime
```

```python
monitoramento = pd.read_csv('monitoramento_tempo.csv')
monitoramento.head()
```

```python
monitoramento.info()
```

```python
monitoramento['data'] = pd.to_datetime(monitoramento['data'], errors = 'coerce')
monitoramento.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize = (16, 8))
plt.grid(True)
plt.plot(monitoramento['data'], monitoramento['temperatura'])
plt.title('Temperatura x Anos')
plt.xlabel('Ano')
plt.ylabel('Temperatura (K)')
```

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(monitoramento['data'], monitoramento['temperatura'], color = 'green')
eixo.set_title('Temperatura x Anos')
eixo.set_xlabel('Ano')
eixo.set_ylabel('Temperatura (K)')
eixo.grid(True)
eixo.legend(['temperatura'], loc = 'lower right', fontsize = 15)
```

## **Customizando as visualizações:**

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(monitoramento['data'], monitoramento['temperatura'], 
          color = 'green',
          linewidth = 1)

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))

eixo.set_title('Temperatura x Anos', fontsize = 15)
eixo.set_xlabel('Ano', fontsize = 15)
eixo.set_ylabel('Temperatura (K)', fontsize = 15)
eixo.grid(True)
eixo.legend(['temperatura'], loc = 'lower right', fontsize = 15)
```

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

eixo.plot(monitoramento['data'], monitoramento['temperatura'], 
          color = 'green',
          linewidth = 1)
eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))

eixo2.plot(monitoramento['data'], monitoramento['temperatura'], 
          color = 'blue',
          linewidth = 1)


eixo.set_title('Temperatura 2014', fontsize = 15)
eixo.set_xlabel('Ano', fontsize = 15)
eixo.set_ylabel('Temperatura (K)', fontsize = 15)
eixo.grid(True)
eixo.legend(['temperatura'], loc = 'upper left', fontsize = 15)
eixo.set_ylim(270, 340)

eixo2.grid(True)
```

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

eixo.plot(monitoramento['data'], monitoramento['temperatura'], 
          color = 'green',
          linewidth = 1)
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1))

eixo2.plot(monitoramento['data'], monitoramento['temperatura'], 
          color = 'blue',
          linewidth = 1)
eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))

eixo.set_title('Temperatura em Maio-2014', fontsize = 15)
eixo.set_xlabel('Ano', fontsize = 15)
eixo.set_ylabel('Temperatura (K)', fontsize = 15)
eixo.grid(True)
eixo.legend(['temperatura'], loc = 'upper left', fontsize = 15)
eixo.set_ylim(270, 320)

eixo2.grid(True)
```

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

eixo.plot(monitoramento['data'], monitoramento['temperatura'], 
          color = 'green',
          linewidth = 1)
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1))

azul_esquerda = monitoramento['data'] < datetime.datetime(2014,5,1)
azul_direito = monitoramento['data'] > datetime.datetime(2014,6,1)

eixo2.plot(monitoramento['data'], monitoramento['temperatura'], 
          color = 'green',
          linewidth = 1)
eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))

eixo2.plot(monitoramento[azul_esquerda]['data'], monitoramento[azul_esquerda]['temperatura'], 
          color = 'blue',
          linewidth = 1)
eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))

eixo2.plot(monitoramento[azul_direito]['data'], monitoramento[azul_direito]['temperatura'], 
          color = 'blue',
          linewidth = 1)
eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))


eixo.set_title('Temperatura em Maio-2014', fontsize = 15)
eixo.set_xlabel('Ano', fontsize = 15)
eixo.set_ylabel('Temperatura (K)', fontsize = 15)
eixo.grid(True)
eixo.legend(['temperatura'], loc = 'upper left', fontsize = 15)
eixo.set_ylim(270, 320)

eixo2.grid(True)
```

## **Anotações:**

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(monitoramento['data'], monitoramento['temperatura'], 
          color = 'green')

eixo.set_title('Temperatura ao Longo do Tempo', fontsize = 15)
eixo.set_xlabel('Ano', fontsize = 15)
eixo.set_ylabel('Temperatura (K)', fontsize = 15)
eixo.grid(True)
eixo.legend(['temperatura'], loc = 'best', fontsize = 15)

x1 = monitoramento['data'][monitoramento['temperatura'].idxmax()]
y1 = max(monitoramento['temperatura'])
eixo.annotate("Máximo", xy = (x1, y1), fontsize = 15, color = 'r')

x2 = monitoramento['data'][monitoramento['temperatura'].idxmin()]
y2 = min(monitoramento['temperatura'])
eixo.annotate("Mínimo", xy = (x2, y2), fontsize = 15, color = 'r')

eixo.axhline(max(monitoramento['temperatura']), color = 'r', ls = '--')
eixo.axhline(min(monitoramento['temperatura']), color = 'r', ls = '--')
```

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(monitoramento['data'], monitoramento['temperatura'], 
          color = 'green')

eixo.set_title('Temperatura ao Longo do Tempo', fontsize = 15)
eixo.set_xlabel('Ano', fontsize = 15)
eixo.set_ylabel('Temperatura (K)', fontsize = 15)
eixo.grid(True)
eixo.legend(['Temperatura'], loc = 'best', fontsize = 15)

x1 = monitoramento['data'][monitoramento['temperatura'].idxmax()]
y1 = max(monitoramento['temperatura'])

x2 = monitoramento['data'][monitoramento['temperatura'].idxmax() - 7000]
y2 = max(monitoramento['temperatura']) - 5

eixo.annotate("Máximo", xy = (x1, y1), fontsize = 15, color = 'r', 
              xytext = (x2, y2),
              arrowprops = dict(facecolor = 'r'))


x1 = monitoramento['data'][monitoramento['temperatura'].idxmin()]
y1 = min(monitoramento['temperatura'])

x2 = monitoramento['data'][monitoramento['temperatura'].idxmin() - 7000]
y2 = min(monitoramento['temperatura']) + 5

eixo.annotate("Mínimo", xy = (x1, y1), fontsize = 15, color = 'r', 
              xytext = (x2, y2),
              arrowprops = dict(facecolor = 'r'))

eixo.axhline(max(monitoramento['temperatura']), color = 'r', ls = '--')
eixo.axhline(min(monitoramento['temperatura']), color = 'r', ls = '--')
```

## **Gráfico de Barras:**

```python
temperatura_por_dia_da_semana = monitoramento.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize = (10, 5))
eixo = fig.add_axes([0, 0, 1, 1])

indice = range(len(temperatura_por_dia_da_semana))

eixo.bar(indice, temperatura_por_dia_da_semana)

eixo.set_title('Temperatura Média na Semana', fontsize = 15)
eixo.set_xlabel('Dia da Semana', fontsize = 15)
eixo.set_ylabel('Temperatura (K)', fontsize = 15)
eixo.grid(True)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize = (10, 5))
eixo = fig.add_axes([0, 0, 1, 1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['k', 'r', 'g', 'b', 'y', 'orange', 'm']

eixo.bar(indice, temperatura_por_dia_da_semana, 
         color = cores, edgecolor = 'k')

eixo.set_title('Temperatura Média na Semana', fontsize = 15)
eixo.set_xlabel('Dia da Semana', fontsize = 15)
eixo.set_ylabel('Temperatura (K)', fontsize = 15)
eixo.grid(True)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

## **Gráfico de Pizza:**

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_da_semana, labels = temperatura_por_dia_da_semana.index)
eixo.set_title('Temperatura Média na Semana', fontsize = 15)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_da_semana, labels = temperatura_por_dia_da_semana.index,
         autopct = '%.1f%%')
eixo.set_title('Temperatura Média na Semana', fontsize = 15)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

fds = [0.1, 0, 0, 0, 0, 0.1, 0.1]

eixo.pie(temperatura_por_dia_da_semana, labels = temperatura_por_dia_da_semana.index,
         autopct = '%.1f%%',
         explode = fds,
         shadow = True)
eixo.set_title('Temperatura Média na Semana', fontsize = 15)
```

## **Gráficos de Dispersão:**

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Gráfico de Dispersão', fontsize = 20)
eixo.set_xlabel('Comprimento da Sépala')
eixo.set_ylabel('Largura da Sépala')
```

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {'Iris-setosa':'r',
        'Iris-versicolor':'b',
        'Iris-virginica': 'g'}

for especie in df['espécie'].unique():
  temp = df[df['espécie'] == especie]
  eixo.scatter(temp['comprimento_sépala'], temp['largura_sépala'], 
               color = cores[especie])

eixo.set_title('Gráfico de Dispersão', fontsize = 20)
eixo.set_xlabel('Comprimento da Sépala')
eixo.set_ylabel('Largura da Sépala')
eixo.legend(cores, fontsize = 15)
```

```python
fig = plt.figure(figsize = (16, 8))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {'Iris-setosa':'r',
        'Iris-versicolor':'b',
        'Iris-virginica': 'g'}

marcadores = {'Iris-setosa':'o',
              'Iris-versicolor':'x',
              'Iris-virginica': 'v'}

for especie in df['espécie'].unique():
  temp = df[df['espécie'] == especie]
  eixo.scatter(temp['comprimento_sépala'], temp['largura_sépala'], 
               color = cores[especie],
               marker = marcadores[especie],
               s = 100)

eixo.set_title('Gráfico de Dispersão', fontsize = 20)
eixo.set_xlabel('Comprimento da Sépala')
eixo.set_ylabel('Largura da Sépala')
eixo.legend(cores, fontsize = 15)
```

```python
fig = plt.figure(figsize = (10, 5))
eixo = fig.add_axes([0, 0, 1, 1])

media, desvpad = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()


eixo.hist(df['comprimento_pétala'],
          edgecolor = 'k', #Cor da Borda
          bins = 20, #Aumenta Número de Caixas
          )
eixo.set_title('Histograma')
eixo.set_xlabel('Comprimento Pétala')
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(media, desvpad),
              xy = (4.5, 20),
              fontsize = 15)
```

```python
fig = plt.figure(figsize = (10, 5))
eixo = fig.add_axes([0, 0, 1, 1])

media, desvpad = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()


eixo.hist(df['comprimento_pétala'],
          edgecolor = 'k', #Cor da Borda
          bins = 20, #Aumenta Número de Caixas
          )
eixo.set_title('Histograma')
eixo.set_xlabel('Comprimento Pétala')
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(media, desvpad),
              xy = (4.5, 20),
              fontsize = 15)

eixo.axvline(media, color = 'r', ls = '--')
eixo.annotate('Média', xy = (media, 28), fontsize = 15, color = 'r')

mediana = df['comprimento_pétala'].median()
eixo.axvline(mediana, color = 'r', ls = '--')
eixo.annotate('Mediana', xy = (mediana, 31), fontsize = 15, color = 'r')
```

```python
fig = plt.figure(figsize = (10, 5))
eixo = fig.add_axes([0, 0, 1, 1])

df_iv = df[df['espécie'] == 'Iris-versicolor']

media, desvpad = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'],
          edgecolor = 'k', #Cor da Borda
          bins = 20, #Aumenta Número de Caixas
          )
eixo.set_title('Iris-versicolor')
eixo.set_xlabel('Comprimento Pétala')
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(media, desvpad),
              xy = (4.6, 6),
              fontsize = 15)

eixo.axvline(media, color = 'r', ls = '--')
eixo.annotate('Média', xy = (media, 5.5), fontsize = 15, color = 'r',
              rotation = 90)

mediana = df['comprimento_pétala'].median()
eixo.axvline(mediana, color = 'r', ls = '--')
eixo.annotate('Mediana', xy = (mediana, 5), fontsize = 15, color = 'r',
              rotation = 90)

fig.savefig('hist_iv', bbox_inches = 'tight')
```
